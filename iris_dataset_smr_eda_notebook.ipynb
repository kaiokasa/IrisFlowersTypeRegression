```python
column = data.columns
for i in column:
    data[i].hist()
    plt.title(i)
    plt.show()
```

Let's rename the target column values with the associated names of the values

```python
dictionary = {0:"setosa",1:"versicolor",2:"virginica"}
data["target_names"] = data["target"].replace(dictionary)
data.head(5)
```

```python
col = data.columns[:-2]
for col in col:
    sns.relplot(x=col, y="target", data=data, hue="target_names")
    
```

```python
sns.pairplot(data, hue="target_names")
```
